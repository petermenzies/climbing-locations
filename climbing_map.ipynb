```python
import pandas as pd
import geopandas as gpd
import folium
```

## Climbing Locations

```python
climbing = pd.read_csv("climbing_locations.csv")
```

```python
geometry = gpd.points_from_xy(x = climbing["long"], y = climbing["lat"])

climbing_geom = climbing.set_geometry(geometry, crs = 4269)
```

```python
point_list = [[point.xy[1][0], point.xy[0][0]] for point in climbing_geom.geometry ]
```

```python
map = folium.Map(tiles = 'Stamen Terrain',
location=[39.84218408207648, -97.59780371796238],
zoom_start=4)
```

```python
for i in range(0, len(point_list)):
    map.add_child(folium.Marker(location = point_list[i],
        tooltip = climbing_geom.name[i],
        popup = climbing_geom.climbing_type[i]+" â€”"+climbing_geom.rock_type[i]))
```
